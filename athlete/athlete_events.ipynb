#### Q2: Find the quantity of gold medals for each type of sport

```python
for i in data.Sport.unique():
    a=data.query(f'Sport=="{i}" and Medal=="Gold"')['Medal'].value_counts()[0]
    print(f'{i}:{a} gold medals')
    print('-'*50)
```

#### Q4: Find the female athletes who competed in winter games and won gold medals in biathlon

```python
f_winner_biathlon_sochi=data.loc[
    (data['Sex']=='F')&
    (data['City']=='Sochi')&
    (data['Sport']=='Biathlon')&
    (data['Medal']=='Gold')
].Name.unique()
f_winner_biathlon_sochi=pd.DataFrame(f_winner_biathlon_sochi, columns=['Name'])
f_winner_biathlon_sochi
```

#### Q4: Find the male athletes who won any medal in summer games and were over the age of 40

```python
male_athletes=data.loc[
    (data['Sex']=='M')&
    (data['Season']=='Summer')&
    (data['Age']>40)
]
medalists=male_athletes[male_athletes['Medal'].notnull()].Name.unique()
df_medalists=pd.DataFrame(medalists, columns=['Name'])
df_medalists
```

#### Q6: Find mean, max, min values for columns Age, Height, Weight

```python
for column in data.iloc[:,[3,4,5]].columns:
    mean=data[column].mean()
    maximum=data[column].max()
    minimum=data[column].min()
    print(f'''Column "{column}":
    \nmean={mean:.1f},
    \nmax={maximum},
    \nmin={minimum}''')
    print('-'*50)
    
```

#### Q6: Find standard deviation and variance for columns Age, Height, Weight

```python
for column in data.iloc[:,[3,4,5]].columns:
    std=data[column].std()
    var=data[column].var()
    print(f'''Column "{column}":
    \nstandard deviation={std:.1f}
    \nvariance={var:.1f}''')
    print('-'*50)
    
```
